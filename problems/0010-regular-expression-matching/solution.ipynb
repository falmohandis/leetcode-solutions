## 1. Problem Summary

We are given an input string `s` and a patern recognizer `p`. We essentially want to see if `s` matches the form of the pattern for `p` in any form.


### What is the input?

`s` is the input string that we are trying to study with this code.
`p` is the pattern we want to try to match to `s` and see if it works or not.

### What is the output?

True or False regarding whether or not there is or is not a match.

### What are the key constraints?
-2^31 <= x <= 2^31 - 1


### Important examples (stolen from their examples section)

```text
Example 1:

Input: x = 121
Output: true
Explanation: 121 reads as 121 from left to right and from right to left.
Example 2:

Input: x = -121
Output: false
Explanation: From left to right, it reads -121. From right to left, it becomes 121-. Therefore it is not a palindrome.
```

````python
import string

class Solution(object):
    def isMatch(self, s, p):
        """
        :type s: str
        :type p: str
        :rtype: bool
        """

        s_list = list(s)
        p_list = list(p)
        abcs = list(string.ascii_lowercase)

        p_pattern = [] # create a list to convert p into an intelligible pattern

        for index, char in enumerate(p_list):  # Trying a very stupid approach to interpret p in a way that allows me to 
            print("Index:", index)
            print("Current pattern character:", char)

            if char == ".":
                p_pattern.append('alphabet')

            elif char == "*":
                previous_char = p_list[index - 1] if index > 0 else None
                if previous_char == ".":
                    previous_char='alphabet'
                p_pattern.append(previous_char)

            elif char.isalpha():
                p_pattern.append(char)

            else:
                raise TypeError("only letters, periods and asterisks can be in the string")

            print("p_pattern is now:")
            print(p_pattern)
            print("-" * 50)

        print("Final p_pattern:")
        print(p_pattern)
        # Check if p_pattern matches s
        if len(s_list) != len(p_pattern):
            print("No match found")
            return False

        for i in range(len(s_list)):
            if p_pattern[i] == 'alphabet' or p_pattern[i] == s_list[i]:
                continue
            else:
                print("No match found")
                return False

        print("Match found")
        return True
````

## 6. Optimized or Alternative Solution

ChatGPT made this for me. It mentioned that this approach simulates the zigzag row by row without building a full 2D grid. Which leads to Time Complexity of O(n) and Space Complexity of O(n).

````python
class OptimizedSolution(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """
        if x < 0:
            return False
        original = x
        reversed_x = 0
        while x > 0:
            reversed_x = reversed_x * 10 + x % 10
            x //= 10
        return original == reversed_x
````

## 7. Test Cases for Optimized Solution

Run the same tests against the optimized solution.

````python
# Manual test cases
def run_my_solution_tests():
    solution = OptimizedSolution() 
    assert solution.isPalindrome(121) == True
    assert solution.isPalindrome(-121) == False
    assert solution.isPalindrome(10) == False
    assert solution.isPalindrome(0) == True
    assert solution.isPalindrome(12321) == True
    assert solution.isPalindrome(12345) == False
    assert solution.isPalindrome(1) == True

    print("All MySolution tests passed.")
run_my_solution_tests()
````
