## 1. Problem Summary

This notebook analyzes LeetCode **Problem 3: Longest Substring Without Repeating Characters**.

The key concept to keep in mind in this problem is understanding the **sliding window** pattern.

A sliding window is a range inside a string that expands and shrinks while you move through the string. For this problem, the window must always contain characters with **no duplicates**.

As we scan the string, we keep track of the characters currently inside the window.


### What is the input?

A string `s`

### What is the output?

Length of the shortest unique consecutive string within `s` which has no repeated letters.

### What are the key constraints?
`s` can have both letters AND numbers.

### Important examples (stolen from their examples section)

```text
Example 1:

Input: s = "abcabcbb"
Output: 3
Explanation: The answer is "abc", with the length of 3. Note that "bca" and "cab" are also correct answers.
Example 2:

Input: s = "bbbbb"
Output: 1
Explanation: The answer is "b", with the length of 1.
Example 3:

Input: s = "pwwkew"
Output: 3
Explanation: The answer is "wke", with the length of 3.
Notice that the answer must be a substring, "pwke" is a subsequence and not a substring.
 
```

## 3. My Original Solutions


## Hack and Slash

````python
class Solution(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        string_library=[]
        dummy_string=[]
        if len(s)==0:
            return 0
        for i in s:
            if i not in dummy_string:
                dummy_string.append(i)
                if i != s[-1]:
                    continue
                else:
                    string_library.append(dummy_string)
                    
            else:
                string_library.append(dummy_string)
                loxn=dummy_string.index(i)
                # Slice from that index to the end
                dummy_string = dummy_string[loxn+1:]
                dummy_string.append(i)
                continue
        try:
            longest = max(string_library, key=len)
        except:
            longest=[i]
        return len(longest)
        

test_string="au"

solution=Solution()
solution.lengthOfLongestSubstring(test_string)
````

````python
class OptimizedSolution(object):
    def lengthOfLongestSubstring(self, s):
        seen = {}
        left = 0
        longest = 0

        for right, char in enumerate(s): 
# Iterate through the string with two pointers, being the left and right pointers.
# The right pointer moves through the string
# The left pointer adjusts to ensure that the substring between them contains unique characters.


            if char in seen and seen[char] >= left: 
# This checks if the character has been seen before and if its last seen index is greater than or equal to the left pointer. 
# If so, it means we have a duplicate character in the current substring.

                left = seen[char] + 1
# Move the left pointer to the right of the last seen index of the duplicate character.

            seen[char] = right
# Update the last seen index of the character to the current index.

            longest = max(longest, right - left + 1)
# Update the longest length found so far by comparing it with the current substring length (right - left + 1).
        return longest
````
