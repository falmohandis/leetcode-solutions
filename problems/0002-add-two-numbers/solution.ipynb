## 1. Problem Summary

We are given two non-empty linked lists. Each linked list represents a non-negative integer, but the digits are stored in **reverse order**.

For example, this linked list:

```text
2 -> 4 -> 3
```

represents the number:

```text
342
```

because the ones digit comes first, then the tens digit, then the hundreds digit.

We need to add the two numbers and return the answer as a linked list, also in reverse order.

### What is the input?

Two linked lists:

```python
l1: Optional[ListNode]
l2: Optional[ListNode]
```

Each node contains one digit from `0` to `9`.

### What is the output?

A linked list representing the sum of the two numbers, with digits stored in reverse order.

### What are the key constraints?

- Each linked list is non-empty.
- Each node stores a single digit from `0` to `9`.
- The numbers do not contain leading zeroes, except the number `0` itself.
- The two linked lists may have different lengths.
- The final answer may need an extra node because of a leftover carry.

### Important examples

```text
Example input:
l1 = 2 -> 4 -> 3
l2 = 5 -> 6 -> 4

Example output:
7 -> 0 -> 8

Explanation:
342 + 465 = 807
Since the answer must be reversed, 807 becomes 7 -> 0 -> 8.
```

```text
Example input:
l1 = 9 -> 9 -> 9 -> 9 -> 9 -> 9 -> 9
l2 = 9 -> 9 -> 9 -> 9

Example output:
8 -> 9 -> 9 -> 9 -> 0 -> 0 -> 0 -> 1

Explanation:
9,999,999 + 9,999 = 10,009,998
Reversed as a linked list, that is 8 -> 9 -> 9 -> 9 -> 0 -> 0 -> 0 -> 1.
```

## 2. Initial Thoughts

At first, this problem can feel strange because the input is not a regular Python list like `[2, 4, 3]`. It is a chain of `ListNode` objects.

A node looks conceptually like this:

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

So this linked list:

```text
2 -> 4 -> 3
```

really means:

```text
Node(val=2, next=Node(val=4, next=Node(val=3)))
```

The solution pattern is:

- Walk through both linked lists at the same time.
- Add the current digit from `l1`, the current digit from `l2`, and the carry.
- Store the ones digit of that sum in a new node.
- Carry the tens digit into the next step.
- Move both pointers forward.

This is very similar to how we add numbers by hand from right to left:

```text
  342
+ 465
-----
  807
```

The reason this problem is convenient is that the digits are already reversed, so the linked list starts at the ones place.

````python
# Imports

from typing import Optional, List


class ListNode:
    """
    Basic linked-list node used by LeetCode.

    A ListNode is one piece of a linked list.
    - val stores the current digit.
    - next points to the next ListNode, or None if this is the last node.
    """
    def __init__(self, val: int = 0, next: Optional["ListNode"] = None):
        self.val = val
        self.next = next

    def __repr__(self):
        return f"ListNode({self.val})"


def build_linked_list(values: List[int]) -> Optional[ListNode]:
    """Convert a Python list like [2, 4, 3] into 2 -> 4 -> 3."""
    dummy = ListNode(0)
    current = dummy

    for value in values:
        current.next = ListNode(value)
        current = current.next

    return dummy.next


def linked_list_to_list(head: Optional[ListNode]) -> List[int]:
    """Convert a linked list like 2 -> 4 -> 3 into a Python list [2, 4, 3]."""
    values = []
    current = head

    while current:
        values.append(current.val)
        current = current.next

    return values


def print_linked_list(head: Optional[ListNode]) -> None:
    """Pretty-print a linked list."""
    values = linked_list_to_list(head)
    print(" -> ".join(map(str, values)) if values else "Empty list")
````

## 3. My Original Solution

Since I was struggling with `ListNode`, a reasonable first solution is to convert the linked lists into normal Python numbers, add them, and then convert the answer back into a linked list.

This is not the most ideal approach for the LeetCode problem, but it is useful for understanding what the problem is asking.

The steps are:

1. Read the first linked list and turn it into an integer.
2. Read the second linked list and turn it into an integer.
3. Add the two integers.
4. Convert the sum back into a reversed linked list.

For example:

```text
2 -> 4 -> 3
```

means:

```text
2 * 1 + 4 * 10 + 3 * 100 = 342
```

````python
class MySolution:
    def linked_list_to_number(self, head: Optional[ListNode]) -> int:
        """
        Convert a reversed linked list into an integer.

        Example:
        2 -> 4 -> 3 becomes 342.
        """
        number = 0
        place_value = 1
        current = head

        while current:
            number += current.val * place_value
            place_value *= 10
            current = current.next

        return number

    def number_to_linked_list(self, number: int) -> ListNode:
        """
        Convert an integer into a reversed linked list.

        Example:
        807 becomes 7 -> 0 -> 8.
        """
        if number == 0:
            return ListNode(0)

        dummy = ListNode(0)
        current = dummy

        while number > 0:
            digit = number % 10
            current.next = ListNode(digit)
            current = current.next
            number //= 10

        return dummy.next

    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        num1 = self.linked_list_to_number(l1)
        num2 = self.linked_list_to_number(l2)
        total = num1 + num2
        return self.number_to_linked_list(total)
````

## 5. Explanation of My Original Solution

### Core idea

Convert both linked lists into regular integers, add the integers, and then build a new linked list from the sum.

### Why it works

Each linked list stores digits in reverse order. That means the first node is the ones place, the second node is the tens place, the third node is the hundreds place, and so on.

For this linked list:

```text
2 -> 4 -> 3
```

we calculate:

```text
2 * 1 + 4 * 10 + 3 * 100 = 342
```

After adding the two numbers, we repeatedly take `number % 10` to get the next digit of the answer.

### Where it may be inefficient

This solution depends on converting the entire linked list into a Python integer. Python can handle very large integers, but this approach misses the main point of the linked-list problem.

In some languages, converting a very large linked list into an integer could overflow the integer type.

### Time complexity

```text
O(max(m, n))
```

where `m` and `n` are the lengths of the two linked lists.

### Space complexity

```text
O(max(m, n))
```

because we create a new linked list for the answer.

## 6. Optimized or Alternative Solution

The standard solution does **not** convert the lists into full numbers.

Instead, it simulates addition digit by digit.

Think about adding these numbers by hand:

```text
  342
+ 465
-----
  807
```

Normally, we add from right to left. Since the linked lists are reversed, the first nodes already represent the rightmost digits.

```text
l1: 2 -> 4 -> 3
l2: 5 -> 6 -> 4
```

Step by step:

```text
2 + 5 = 7        write 7, carry 0
4 + 6 = 10       write 0, carry 1
3 + 4 + 1 = 8    write 8, carry 0
```

Answer:

```text
7 -> 0 -> 8
```

The clean trick is to use a `dummy` node. The dummy node is not part of the real answer. It just gives us an easy starting point so we do not need special logic for the first node.

````python
class OptimizedSolution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode(0)
        current = dummy
        carry = 0

        # Continue while either list still has nodes or there is a leftover carry.
        while l1 or l2 or carry:
            val1 = l1.val if l1 else 0
            val2 = l2.val if l2 else 0

            total = val1 + val2 + carry
            digit = total % 10
            carry = total // 10

            current.next = ListNode(digit)
            current = current.next

            if l1:
                l1 = l1.next
            if l2:
                l2 = l2.next

        return dummy.next
````

## 9. Pattern Recognition

### Pattern used in this problem

This problem uses the **linked list traversal** pattern plus a simple **math simulation** pattern.

We move through two linked lists one node at a time:

```python
if l1:
    l1 = l1.next
if l2:
    l2 = l2.next
```

At each step, we simulate one column of addition:

```python
total = val1 + val2 + carry
digit = total % 10
carry = total // 10
```

### How I can recognize this pattern next time

Use this pattern when:

- The problem gives the head of a linked list.
- You need to process nodes from beginning to end.
- You see `.next` in the data structure.
- You need to build a new linked list as the answer.
- A problem asks you to simulate a process one node or one digit at a time.

Important linked-list idea:

```text
head
 ↓
[2] -> [4] -> [3] -> None
```

`head` is not the whole list as an array. It is a reference to the first node. From there, each node points to the next one.

## 10. Edge Cases

| Edge Case | Expected Behavior | Covered? |
|---|---|---|
| `l1 = [0]`, `l2 = [0]` | Return `[0]` | Yes |
| Lists have different lengths | Treat missing digits as `0` | Yes |
| Final leftover carry | Add one extra node | Yes |
| Long chain of carry values, like `[9,9,9] + [1]` | Keep carrying until finished | Yes |
| One list ends before the other | Continue processing the remaining list | Yes |

The optimized solution handles these because the loop condition is:

```python
while l1 or l2 or carry:
```

That means we keep going as long as there is still a node in either list or a carry that has not been written yet.

## 11. Mistakes, Bugs, or Confusions

### Mistakes

The biggest confusion was understanding what a `ListNode` actually is.

A Python list is like this:

```python
[2, 4, 3]
```

A linked list is more like this:

```text
ListNode(2) -> ListNode(4) -> ListNode(3) -> None
```

Each node only knows two things:

```python
node.val   # the value inside this node
node.next  # the next node
```

Common mistakes in this problem:

- Thinking `l1` is a full Python list.
- Forgetting to move forward with `l1 = l1.next` and `l2 = l2.next`.
- Forgetting to include the carry in the next addition.
- Forgetting to create a final node when carry remains at the end.
- Returning `dummy` instead of `dummy.next`.

### How I fixed them

I used a `dummy` node and a `current` pointer:

```python
dummy = ListNode(0)
current = dummy
```

Then every time I calculate a new digit, I attach it to the answer:

```python
current.next = ListNode(digit)
current = current.next
```

At the end, I return:

```python
return dummy.next
```

because `dummy` is just a fake starter node, not part of the actual answer.

## 12. Final Clean Solution

This is the final version I would submit to LeetCode.

LeetCode already provides the `ListNode` class, so the submission usually only needs the `Solution` class.

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode(0)
        current = dummy
        carry = 0

        while l1 or l2 or carry:
            val1 = l1.val if l1 else 0
            val2 = l2.val if l2 else 0

            total = val1 + val2 + carry
            carry = total // 10
            digit = total % 10

            current.next = ListNode(digit)
            current = current.next

            if l1:
                l1 = l1.next
            if l2:
                l2 = l2.next

        return dummy.next
````

## 13. Final Notes

Because the digits are stored in reverse order, we can add from the head of each linked list directly. The head node is the ones place, which is exactly where manual addition starts.

### What I learned

A `ListNode` is a small object that stores a value and a reference to the next node. A linked list is created by chaining nodes together.

The line:

```python
current.next = ListNode(digit)
```

adds a new node after the current node.

```python

```
