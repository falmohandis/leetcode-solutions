## 1. Problem Summary

In this problem, we want to check if a number is a palindrome or not.


### What is the input?

An integer called `x` which is in the range of -231 <= x <= 231 - 1. 

### What is the output?

True or False

### What are the key constraints?
-2^31 <= x <= 2^31 - 1


### Important examples (stolen from their examples section)

```text
Example 1:

Input: x = 121
Output: true
Explanation: 121 reads as 121 from left to right and from right to left.
Example 2:

Input: x = -121
Output: false
Explanation: From left to right, it reads -121. From right to left, it becomes 121-. Therefore it is not a palindrome.
Example 3:

Input: x = 10
Output: false
Explanation: Reads 01 from right to left. Therefore it is not a palindrome.
 
```

## 3. My Original Solutions


## Hack and Slash

````python
class Solution(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """
        if x<0:
            return False
        digits = [int(d) for d in str(x)]
        is_pal = digits == digits[::-1]
        return is_pal
````

## 6. Optimized or Alternative Solution

ChatGPT made this for me. It mentioned that this approach simulates the zigzag row by row without building a full 2D grid. Which leads to Time Complexity of O(n) and Space Complexity of O(n).

````python
class OptimizedSolution(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """
        if x < 0:
            return False
        original = x
        reversed_x = 0
        while x > 0:
            reversed_x = reversed_x * 10 + x % 10
            x //= 10
        return original == reversed_x
````
