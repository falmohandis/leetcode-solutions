## 1. Problem Summary

The problem itself is straightforward- you get an array (in this case a list) of integers as well as a target value that is the sum of two values in the list. Thinking out loud, there is a very engineer-thinking mentality brute force solution, which would be to use a wide array of loops, which is largely inefficient and typical of a person who cares very little about efficiency... aka me before starting this journey with LeetCode. I will demonstrate this method, my own method post learning more about datatypes and thereafter what GPT gives to see what is the best approach.

### What is the input?

A list called `nums` with a variety of integers and a target integer called `target` which demonstrates the goal sum.

### What is the output?

Indices of the two values from `nums` which add up to make `target`. 

### What are the key constraints?
The order can technically be in any order, but the same element cannot be used twice. There will be exactly 1 solution.

## 3. My Original Solutions


## Hack and Slash

````python
# Really dumb engineering solution without the use of "hash maps" aka "dictionaries."
class Solution(object):
    def twoSum(self, nums, target):
        for i in range(len(nums)):
            dummy_target = target - nums[i]

            dummy_nums = nums[:]
            dummy_nums.pop(i)

            if dummy_target in dummy_nums:
                location = nums.index(dummy_target, i + 1)
                return [i, location]
````

## 6. Optimized or Alternative Solution

Use this section for a cleaner, faster, or more standard solution.

This can be your improved solution, a solution learned from discussion, or a comparison solution generated analytically.

````python
class OptimizedSolution:
    def twoSum(self,nums, target):
        seen = {}

        for i, num in enumerate(nums):
            complement = target - num

            if complement in seen:
                return [seen[complement], i]

            seen[num] = i

        
````
