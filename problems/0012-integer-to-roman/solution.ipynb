## 1. Problem Summary

Given an integer, we want to convert it into roman numerals.


### What is the input?

Integer `num` is the input. I need to code in the conversion from integer to roman numeral.

### What is the output?

A string which is the roman numeral

### What are the key constraints?
1 <= num <= 3999


### Important examples (stolen from their examples section)

```text
Example 1:

Input: num = 3749

Output: "MMMDCCXLIX"

Explanation:

3000 = MMM as 1000 (M) + 1000 (M) + 1000 (M)
 700 = DCC as 500 (D) + 100 (C) + 100 (C)
  40 = XL as 10 (X) less of 50 (L)
   9 = IX as 1 (I) less of 10 (X)
Note: 49 is not 1 (I) less of 50 (L) because the conversion is based on decimal places
Example 2:

Input: num = 58

Output: "LVIII"

Explanation:

50 = L
 8 = VIII
Example 3:

Input: num = 1994

Output: "MCMXCIV"

Explanation:

1000 = M
 900 = CM
  90 = XC
   4 = IV
   
```

But...this didnt work in all cases because of some minor technicalities with numbers that become IV and IX... so I had GPT edit my solution slightly.

````python
class Solution(object):
    def intToRoman(self, num):
        """
        :type num: int
        :rtype: str
        """

        roman_output = ''
        dummy_num = num

        while dummy_num >= 1000:
            dummy_num -= 1000
            roman_output += 'M'

        while dummy_num >= 900:
            dummy_num -= 900
            roman_output += 'CM'

        while dummy_num >= 500:
            dummy_num -= 500
            roman_output += 'D'

        while dummy_num >= 400:
            dummy_num -= 400
            roman_output += 'CD'

        while dummy_num >= 100:
            dummy_num -= 100
            roman_output += 'C'

        while dummy_num >= 90:
            dummy_num -= 90
            roman_output += 'XC'

        while dummy_num >= 50:
            dummy_num -= 50
            roman_output += 'L'

        while dummy_num >= 40:
            dummy_num -= 40
            roman_output += 'XL'

        while dummy_num >= 10:
            dummy_num -= 10
            roman_output += 'X'

        while dummy_num >= 9:
            dummy_num -= 9
            roman_output += 'IX'

        while dummy_num >= 5:
            dummy_num -= 5
            roman_output += 'V'

        while dummy_num >= 4:
            dummy_num -= 4
            roman_output += 'IV'

        while dummy_num >= 1:
            dummy_num -= 1
            roman_output += 'I'

        return roman_output
````

## 6. Optimized or Alternative Solution

ChatGPT made this for me using a hash table.

````python
class OptimizedSolution(object):
    def intToRoman(self, num):
        """
        :type num: int
        :rtype: str
        """

        thousands = {
            0: "",
            1: "M",
            2: "MM",
            3: "MMM"
        }

        hundreds = {
            0: "",
            1: "C",
            2: "CC",
            3: "CCC",
            4: "CD",
            5: "D",
            6: "DC",
            7: "DCC",
            8: "DCCC",
            9: "CM"
        }

        tens = {
            0: "",
            1: "X",
            2: "XX",
            3: "XXX",
            4: "XL",
            5: "L",
            6: "LX",
            7: "LXX",
            8: "LXXX",
            9: "XC"
        }

        ones = {
            0: "",
            1: "I",
            2: "II",
            3: "III",
            4: "IV",
            5: "V",
            6: "VI",
            7: "VII",
            8: "VIII",
            9: "IX"
        }

        return (
            thousands[num // 1000] +
            hundreds[(num % 1000) // 100] +
            tens[(num % 100) // 10] +
            ones[num % 10]
        )
````
