## 1. Problem Summary

In this problem, we essentially want to reverse a number while preserving its sign. 


### What is the input?

An integer called `x` which is in the range of -231 <= x <= 231 - 1. 

### What is the output?

The output is a reversed order of the values while the sign is kept.


### What are the key constraints?
-231 <= x <= 231 - 1

## 3. My Original Solutions


## Hack and Slash

````python
class Solution(object):
    def myAtoi(self, s):
        s = list(s)

        output = []
        sign = 1

        while s and s[0] == " ":
            del s[0]

        if s and s[0] == '-':
            sign = -1
            del s[0]

        elif s and s[0] == '+':
            del s[0]

        print(s)

        for i in s:
            if i.isalpha():
                if output == []:
                    return 0
                else:
                    result = int("".join(map(str, output))) * sign

                    if result < -2**31:
                        return -2**31
                    elif result > 2**31 - 1:
                        return 2**31 - 1

                    return result

            elif i.isdigit():
                output.append(i)
                continue

            else:
                break

        if output == []:
            return 0

        result = int("".join(map(str, output))) * sign

        if result < -2**31:
            return -2**31
        elif result > 2**31 - 1:
            return 2**31 - 1

        return result
````

````python
# Manual test cases
def run_my_solution_tests():
    solution = Solution() 
    assert solution.myAtoi("42") == 42
    assert solution.myAtoi("   -42") == -42
    assert solution.myAtoi("4193 with words") == 4193
    assert solution.myAtoi("words and 987") == 0
    assert solution.myAtoi("-91283472332") == -2147483648
    assert solution.myAtoi("3.14159") == 3
    assert solution.myAtoi("+1") == 1
    assert solution.myAtoi("+-2") == 0
    assert solution.myAtoi("   +0 123") == 0
    assert solution.myAtoi("") == 0
    assert solution.myAtoi("   ") == 0
    assert solution.myAtoi("0") == 0
    assert solution.myAtoi("-") == 0
    assert solution.myAtoi("+") == 0

    print("All MySolution tests passed.")
run_my_solution_tests()
````
