## 3. My Original Solutions


## Hack and Slash

````python
import numpy as np

class Solution(object):
    def convert(self, s, numRows):
        """
        :type s: str
        :type numRows: int
        :rtype: str
        """

        if numRows == 1 or numRows >= len(s):
            return s

        dummy_array = np.full((numRows, len(s)), '', dtype=str)

        row = 0
        column = 0

        up = False
        down = True

        for ch in s:

            dummy_array[row, column] = ch

            if down:

                if row == numRows - 1:
                    down = False
                    up = True
                    row -= 1
                    column += 1
                else:
                    row += 1

            else:  # moving up diagonally

                if row == 0:
                    down = True
                    up = False
                    row += 1
                else:
                    row -= 1
                    column += 1

        return "".join(
            "".join(item for item in row if item)
            for row in dummy_array
        )
````

## 4. Test Cases for My Original Solution


````python
# Manual test cases
def run_my_solution_tests():
    solution = Solution() 
    assert solution.convert("PAYPALISHIRING", 3) == "PAHNAPLSIIGYIR"
    assert solution.convert("PAYPALISHIRING", 4) == "PINALSIGYAHRPI"
    assert solution.convert("A", 1) == "A"
    
    print("All MySolution tests passed.")
run_my_solution_tests()
````

## 6. Optimized or Alternative Solution

ChatGPT made this for me. It mentioned that this approach simulates the zigzag row by row without building a full 2D grid. Which leads to Time Complexity of O(n) and Space Complexity of O(n).

````python
class OptimizedSolution:
    def convert(self, s: str, numRows: int) -> str:
        if numRows == 1 or numRows >= len(s):
            return s

        rows = [""] * numRows
        curr_row = 0
        going_down = False

        for char in s:
            rows[curr_row] += char

            if curr_row == 0 or curr_row == numRows - 1:
                going_down = not going_down

            curr_row += 1 if going_down else -1

        return "".join(rows)
````

## 7. Test Cases for Optimized Solution
