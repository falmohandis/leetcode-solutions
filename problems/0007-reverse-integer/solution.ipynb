## 1. Problem Summary

In this problem, we essentially want to reverse a number while preserving its sign. 


### What is the input?

An integer called `x` which is in the range of -231 <= x <= 231 - 1. 

### What is the output?

The output is a reversed order of the values while the sign is kept.


### What are the key constraints?
-231 <= x <= 231 - 1

## 3. My Original Solutions


## Hack and Slash

````python
import numpy as np

class Solution(object):
    def reverse(self, x):
        """
        :type x: int
        :rtype: int
        """
        if x<0:
            return -self.reverse(-x)
        if abs(x) > 2**31-1:
            return 0
        x=list(map(int, str(x)))
        x.reverse()
        res=int("".join(map(str, x)))
        return res
````

## 4. Test Cases for My Original Solution


````python
# Manual test cases
def run_my_solution_tests():
    solution = Solution() 
    assert solution.reverse(123) == 321
    assert solution.reverse(-123) == -321
    assert solution.reverse(120) == 21
    assert solution.reverse(0) == 0

    print("All MySolution tests passed.")
run_my_solution_tests()
````
